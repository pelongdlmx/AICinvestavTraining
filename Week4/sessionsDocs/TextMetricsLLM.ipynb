```python
import argparse
import torch
from transformers import GPT2LMHeadModel, GPT2TokenizerFast

def load_model_and_tokenizer(device: str = "cpu"):
    """
    Load the pretrained GPT‑2 small checkpoint and its tokenizer.
    """
    # Tokenizer: converts raw text ↔ token IDs
    tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")  # 117M params
    tokenizer.pad_token = tokenizer.eos_token   # required for padding in generation

    # Model: causal LM head
    model = GPT2LMHeadModel.from_pretrained("gpt2")
    model.to(device)
    model.eval()   # set to inference mode (turn off dropout, etc.)
    return tokenizer, model

def generate_text(
    prompt: str,
    tokenizer,
    model,
    device: str = "cpu",
    max_new_tokens: int = 10,
    temperature: float = 1.0,
    top_k: int = 50,
    top_p: float = 0.95,
    repetition_penalty: float = 1.0,
):
    """
    Generate a continuation of `prompt` using GPT‑2.
    """
    # Encode the prompt
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)

    # Generate tokens
    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            max_new_tokens=max_new_tokens,
            temperature=temperature,
            top_k=top_k,
            top_p=top_p,
            repetition_penalty=repetition_penalty,
            pad_token_id=tokenizer.eos_token_id,
        )

    # Decode back to text
    generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return generated_text
```

```python
tokenizer, model = load_model_and_tokenizer()
```

```python
generate_text("hi", tokenizer, model)
```
